### Generating Equilibrium Prices

Simulate some prices from a Bertrand Nash Equilibrium.  

```julia
cd("/Users/VGupta/Documents/Research/Julia Stuff")
include("BertrandNash.jl");  # does the generation of the prices
include("fitNash.jl");  # does the fitting
```

```julia
### Parameters
using PyPlot
srand(8675309)
thetas1 = [-1, .3]
thetas2 = [.5, -1.2]
pmax = 5;
N = 200;
```

Now iterate through, generate prices and the price and residuals array

```julia
inPrices = Array(Float64, N, 2)
errors = Array(Float64, N, 3)
for iRun = 1:N
	intercept1, intercept2, totNoise = genErrors()
	pstar = solveNashPrices(thetas1, thetas2, intercept1, intercept2, pmax)
	inPrices[iRun, :] = pstar'
    errors[iRun, :] = [intercept1, intercept2, totNoise]
end
```

```julia
using PyPlot
subplot(1, 3, 1)
scatter(inPrices[:, 1], inPrices[:, 2], color="blue")
title("Price Scatter Plot")
subplot(1, 3, 2)
scatter(errors[:, 3], inPrices[:, 1], color="red")
title("Price 1 by GDP")

subplot(1, 3, 3)
scatter(errors[:, 3], inPrices[:, 2], color="red")
title("Price 2 by GDP")

```

### Fitting    

```julia
m, vthetas1, vthetas2, vintercept1, vintercept2 = setUpFitting()
resids = Variable[]
@defVar(m, maxVal >=0)
for iRun = 1:N
    resid = addResids(m, vthetas1, vthetas2, vintercept1, vintercept2, inPrices[iRun, :], errors[iRun, 3], pmax)
	push!(resids, resid)
	@addConstraint(m, maxVal >= resid)
end
@setObjective(m, Min, sum{resids[i], i=1:N}/N)
solve(m)
```

```julia
fTheta1 = [getValue(vthetas1[i]) for i =1:3]
fTheta2 = [getValue(vthetas2[i]) for i = 1:3]
fIntercept1 = getValue(vintercept1)
fIntercept2 = getValue(vintercept2)
```

### Out of Sample Analysis

```julia
resid_vals = [getValue(resids[i])::Float64 for i = 1:N ]
srand(516746226)
fTheta1
```

```julia
solveNashPricesFit(thetas1, thetas2, intercept1, intercept2, GDP, pmax) = 
	solveNashPrices(thetas1[1:2], thetas2[1:2], thetas1[3] * GDP + intercept1, thetas2[3] * GDP + intercept2, pmax)
```

```julia
Nout = 200;
fPrices = Array(Float64, Nout, 2)
oPrices = Array(Float64, Nout, 2)
resid_out = Array(Float64, Nout)
oErrors = Array(Float64, Nout, 3)
brPrices = Array(Float64, Nout, 2)

for iRun = 1:Nout
	intercept1, intercept2, GDP = genErrors()
    oErrors[iRun, :] = [intercept1, intercept2, GDP]
    pfit = solveNashPricesFit(fTheta1, fTheta2, fIntercept1, fIntercept2, GDP, pmax)
    pstar = solveNashPrices(thetas1, thetas2, intercept1, intercept2, pmax)

    fPrices[iRun, :] = pfit'
    oPrices[iRun, :] = pstar'
    resid_out[iRun] = calcResid(fTheta1, fTheta2, fIntercept1, fIntercept2, pstar, GDP, pmax)
    brPrices[iRun, 1] = bestResponsePrice(fTheta1[1:2], fIntercept1 + fTheta1[3] * GDP, pstar, pmax, :1)
    brPrices[iRun, 2] = bestResponsePrice(fTheta2[1:2], fIntercept2 + fTheta2[3] * GDP, pstar, pmax, :2)

end
```

Notice that out of sample performance looks a lot like in sample performance.

```julia
subplot(1, 2, 1)
plt.hist(resid_vals, 20, normed=true)
title("In Sample Residuals")
text(1, 1,  "mean = $(round(mean(resid_vals), 3)), \n sd = $(round(std(resid_vals), 3))")

subplot(1, 2, 2)
plt.hist(resid_out, 20, normed=true)
title("OutSample Resids")
text(1.4, 1,  "mean = $(round(mean(resid_out), 3)), \n sd = $(round(std(resid_out), 3))")
tight_layout()
```

Look at the histogram of differences in prices...

```julia
subplot(1, 2, 1)
price_diffs = oPrices[:, 1] - fPrices[:, 1]
plt.hist(price_diffs, 20, normed=true, color="green")
title("Price 1 diffs")
text(.05, 3,  "mean = $(round(mean(price_diffs), 3)), \n sd = $(round(std(price_diffs), 3))")

subplot(1, 2, 2)
price_diffs = oPrices[:, 2] - fPrices[:, 2]
plt.hist(price_diffs, 20, normed=true, color="green")
title("Price 2 diffs")
text(.05, 3,  "mean = $(round(mean(price_diffs), 3)), \n sd = $(round(std(price_diffs), 3))")
```
