### Generating Equilibrium Prices

Simulate some prices from a Bertrand Nash Equilibrium.  Notice using v2 of these files...

```julia
using PyPlot, Distributions, DataFrames
cd("/Users/VGupta/Documents/Research/Julia Stuff/Demand")
include("BertrandNash2.jl");  # does the generation of the prices
include("fitNash2.jl");  # does the fitting
```

```julia
### Parameters
thetas1 = [-1.2, .5, 1, -10]
thetas2 = [.3, -1, 1, -10]
pmax = .45; N = 200;

function genErrors( ; mu1 = 5, mu2 = 5, mu3 = 5, sigma1 = 1.5, sigma2 = 1.5, sigma3 = 1.5)
    eps1 = rand(Normal(mu1, sigma1)); eps2 = rand(Normal(mu2, sigma2)); eps3 = rand(Normal(mu3, sigma3))
    eps3 = (eps1 + eps2 + eps3)/3
	return eps1, eps2, eps3
end
```

Now iterate through, generate prices and the price and residuals array

```julia
srand(8675309); inPrices = Array(Float64, N, 2); inErrors = Array(Float64, N, 3)
for iRun = 1:N
    inErrors[iRun, :] = [genErrors()...]
    pstar = solveNashPrices(thetas1, thetas2, inErrors[iRun, :], pmax, prices=[.3, .3])
	inPrices[iRun, :] = pstar'
end
indx = find( abs(inErrors[:, 3] - median(inErrors[:, 3]) ) .<= 1e-2 )[1];
```

### Fitting    

```julia
#Try thinning out p1_grid, p2_grid
function thin(p1_grid, delta, pmax)
    p1_gridp = Float64[]; push!(p1_gridp, p1_grid[1])
    for i = 2:N
        if p1_grid[i] - p1_gridp[length(p1_gridp)] > delta
            push!(p1_gridp, p1_grid[i] )
        end
    end
    if abs(p1_gridp[length(p1_gridp)] - pmax) > .05
        push!(p1_gridp, pmax)
    end
    return p1_gridp
end
```

```julia
#build a kernel, the graham matrix, and it's cholesky
c = .01
lambda_reg = 1e-7

k(x,y) = exp( -1/c * norm(x-y)^2 )
data = [inPrices inErrors[:, 3]]
p1_grid = sort(inPrices[:, 1]); p2_grid = sort(inPrices[:, 2])

### Thin out the grids
# p1_grid = thin(p1_grid, .45/50, pmax)
# p2_grid = thin(p2_grid, .45/50, pmax)
p1_grid = p2_grid = linspace(.1, pmax, 50)

N1 = length(p1_grid); N2 = length(p2_grid); totN = N + N1 + N2
data = [data; p1_grid repmat([inPrices[indx, 2]], N1, 1) repmat([inErrors[indx, 3]], N1, 1)]
data = [data; repmat([inPrices[indx, 1]], N2, 1) p2_grid repmat([inErrors[indx, 3]], N2, 1)]
data = [data; p1_grid[1] inPrices[indx, 2] inErrors[indx, 3];
              inPrices[indx, 1] p2_grid[1] inErrors[indx, 3]]

K = [k(data[i, :], data[j, :]) for i =1:N+N1+N2+2, j=1:totN+2]  #Try for now to do it at scale without trimming
D, V = eig(K)

#only use significant eigenvalues
tailN = find( 1 - (cumsum(D) / sum(D)) .<= 1-1e-6)[1]

Dbar = copy(D)
Dbar[1:tailN-1] = 0;
Vbar = V * diagm(sqrt(Dbar));

m, vZ1, vZ2, vf1, vf2, vIntercept = setUpFitting(totN+2, Vbar[:, tailN:totN+2])

#Right now we're normalizing at the minimum
normalize(m, vf1[totN + 1], margRev(thetas1, data[totN + 1, 1:2], data[totN + 1, 3], :1))                
normalize(m, vf2[totN + 2], margRev(thetas2, data[totN + 2, 1:2], data[totN + 2, 3], :2))                

#Alternate normalization procedure... normalize the average value of the function over the grid
# println("Mean Value 1:\t", mean([margRev(thetas1, data[ix, 1:2], data[ix, 3], :1) for ix = N+1:N+N1] ) )
# println("Mean Value 2:\t", mean([margRev(thetas2, data[ix, 1:2], data[ix, 3], :1) for ix = N+N1+1:totN] ) )
# @addConstraint(m, sum{vf1[ix], ix=N+1:N+N1} == sum([margRev(thetas1, data[ix, 1:2], data[ix, 3], :1) for ix = N+1:N+N1]))
# @addConstraint(m, sum{vf2[ix], ix=N+N1+1:totN} == sum([margRev(thetas2, data[ix, 1:2], data[ix, 3], :1) for ix = N+N1+1:totN]))


resids = Variable[]
@defVar(m, maxVal >=0)
for iRun = 1:N
    resid = addResids(m, vf1, vf2, iRun, data, pmax)
    push!(resids, resid)
	@addConstraint(m, maxVal >= resid)
end

addDecreasing(m, vf1, (N+1):N+N1, TOL=1e-6 )
addDecreasing(m, vf2, (N+N1+1):totN, TOL=1e-6 )
addPositivity(m, vf1, totN + 1)
addPositivity(m, vf2, totN + 2)

@defVar(m, reg_term[1:2])
HNorm1 = QuadExpr(vZ1[:], vZ1[:], ones(length(vZ1)), AffExpr())
HNorm2 = QuadExpr(vZ2[:], vZ2[:], ones(length(vZ2)), AffExpr())
addConstraint(m, HNorm1 <= reg_term[1]) 
addConstraint(m, HNorm2 <= reg_term[2])

# @setObjective(m, Min, maxVal + lambda_reg * (reg_term[1] + reg_term[2]))
@setObjective(m, Min, sum{resids[i], i=1:N}/N + lambda_reg * (reg_term[1] + reg_term[2]))

# @setObjective(m, Min, maxVal)
# @setObjective(m, Min, sum{resids[i], i=1:N}/N)
println(solve(m))
getObjectiveValue(m)

norm1 = getValue(reg_term[1]); norm2 = getValue(reg_term[2])
target_val = getObjectiveValue(m) - lambda_reg * (norm1 + norm2) 

target_val, norm1, norm2

# # #resolve once more to try and get resids btter
# # @addConstraint(m, maxVal + lambda_reg * reg_term <= getObjectiveValue(m) + 1e-6)
# # @setObjective(m, Min, sum{resids[i], i=1:N})
# # println(solve(m))

```

### Out of Sample Analysis

```julia
Nout = 200;
srand(516746226)

fPrices = Array(Float64, Nout, 2)
oPrices = Array(Float64, Nout, 2)
resid_out = Array(Float64, Nout)
brPrices = Array(Float64, Nout, 2)
oErrors = Array(Float64, Nout, 3)
for iRun = 1:Nout
    oErrors[iRun, :] = [genErrors()...]
    pstar = solveNashPrices(thetas1, thetas2, oErrors[iRun, :], pmax)
    oPrices[iRun, :] = pstar'
    fPrices[iRun, :] = solveNashPrices(fMR1_fun, fMR2_fun, oErrors[iRun, 3], pmax, 
                                prices=copy(pstar), trace=false)
    resid_out[iRun] = calcResid(fMR1_fun, fMR2_fun, [pstar' oErrors[iRun, 3]], pmax)
end
```

### Ambiguity Sets

```julia
alpha_grid = -9.:1.5:0
m, vlambdas1, vlambdas2, features = setUpFitting(alpha_grid)

#Right now match at the mode(myPrices) median values of other variables
normalize(m, vlambdas1, features, [minimum(inPrices[:, 1]), inPrices[indx, 2]], inErrors[indx, 3], 
                   margRev(thetas1, [minimum(inPrices[:, 1]), inPrices[indx, 2]], inErrors[indx, 1], :1) )
normalize(m, vlambdas2, features, [inPrices[indx, 1], minimum(inPrices[:, 2])], inErrors[indx, 3], 
                   margRev(thetas2, [inPrices[indx, 1], minimum(inPrices[:, 2])], inErrors[indx, 2], :2) )

resids = Variable[]
@defVar(m, 0 <= maxVal <= target_val)  #constrain that we can't be too far away
for iRun = 1:N
    resid = addResids(m, vlambdas1, vlambdas2, inPrices[iRun, :], inErrors[iRun, 3], pmax, features)
    push!(resids, resid)
	@addConstraint(m, maxVal >= resid)
end
numFeatures = length(vlambdas1)
addPositivity(m, vlambdas1, features, inPrices[indx, 2], minimum(inPrices[:, 1]), inErrors[indx, 3], :1)
addPositivity(m, vlambdas2, features, inPrices[indx, 1], minimum(inPrices[:, 2]), inErrors[indx, 3], :2)

addDecreasingCnsts(m, vlambdas1, features, inPrices[indx, 2], inErrors[indx, 3], inPrices[:, 1], pmax, :1)
addDecreasingCnsts(m, vlambdas2, features, inPrices[indx, 1], inErrors[indx, 3], inPrices[:, 2], pmax, :2)

# #Constrain the 1 norm of the parameters to prevent wildness
@defVar(m, t1[1:numFeatures] >= 0)
@defVar(m, t2[1:numFeatures] >= 0)
for i =1:numFeatures
    @addConstraint(m, t1[i] >= vlambdas1[i])
    @addConstraint(m, t1[i] >= -vlambdas1[i])
    @addConstraint(m, t2[i] >= vlambdas2[i])
    @addConstraint(m, t2[i] >= -vlambdas2[i])
end
@defVar(m, reg_term <= ball_size)

case = :Sample1    
if case == :Max1
    for j = 1:N
        bounds1[j, 1] = fMR1_fun([p1_grid[j], inPrices[indx, 2]], inErrors[indx, 3])
        @setObjective(m, Max,  
            sum{features[i](p1_grid[j], inPrices[indx, 2], inErrors[indx, 3])*vlambdas1[i], i=1:numFeatures}) 
        status = solve(m)
        if status != :Optimal 
            println("Warning: $status")
        end
        bounds1[j, 3] = getObjectiveValue(m)
    end
elseif case == :Min1
    for j = 1:N
        @setObjective(m, Min,  
            sum{features[i](p1_grid[j], inPrices[indx, 2], inErrors[indx, 3])*vlambdas1[i], i=1:numFeatures}) 
        status = solve(m)
        if status != :Optimal 
            println("Warning: $status")
        end
        bounds1[j, 2] = getObjectiveValue(m)
    end
elseif case ==:Max2    
    for j = 1:N
        bounds2[j, 1] = fMR2_fun([inPrices[indx, 1], p2_grid[j]], inErrors[indx, 3])
        @setObjective(m, Max,  
            sum{features[i](inPrices[indx, 1], p2_grid[j], inErrors[indx, 3])*vlambdas2[i], i=1:numFeatures}) 
        status = solve(m)
        if status != :Optimal 
            println("Warning: $status")
        end
        bounds2[j, 3] = getObjectiveValue(m)
    end
elseif case == :Min2
    for j = 1:N
        @setObjective(m, Min,  
            sum{features[i](inPrices[indx, 1], p2_grid[j], inErrors[indx, 3])*vlambdas2[i], i=1:numFeatures}) 
        status = solve(m)
        if status != :Optimal 
            println("Warning: $status")
        end
        bounds2[j, 2] = getObjectiveValue(m)
    end
elseif case == :Sample1
    @setObjective(m, Min,  
        sum{features[i](p1_grid[25], inPrices[indx, 2], inErrors[indx, 3])*vlambdas1[i], i=1:numFeatures}) 
    status = solve(m)
    if status != :Optimal 
        println("Warning: $status")
    end
    
    flambdas1 = [getValue(vlambdas1[i]) for i =1:length(vlambdas1)]
    fMR1_fun = createfun(flambdas1, features)
    for j = 1:N
        bounds1[j, 4] = fMR1_fun([p1_grid[j], inPrices[indx, 2]], inErrors[indx, 3])
    end
elseif case == :Sample2
    @setObjective(m, Min,  
            sum{features[i](inPrices[indx, 1], p2_grid[25], inErrors[indx, 3])*vlambdas2[i], i=1:numFeatures}) 
    status = solve(m)
    if status != :Optimal 
        println("Warning: $status")
    end
    flambdas2 = [getValue(vlambdas2[i]) for i = 1:length(vlambdas1)]
    fMR2_fun = createfun(flambdas2, features)
    for j = 1:N
        bounds2[j, 4] = fMR2_fun([inPrices[indx, 1], p2_grid[j]], inErrors[indx, 3])
    end
end
```
